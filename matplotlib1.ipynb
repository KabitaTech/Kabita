```python
##importing the library

import numpy as np
import matplotlib.pyplot as plt

%matplotlib  inline
#defining the dataset
x=np.arange(0,10,2)
y=3*x+5

fig=plt.figure(figsize=(10,5))
#plotting the datapoints
plt.plot(x,y,linewidth =2.0,linestyle= ":", color='b',alpha= 0.5, marker='o')
plt.title("LINE X/Y Axis PLOT" )
plt.xlabel("X-AXIS ")
plt.ylabel("Y-AXIS ")
plt.legend(['line1'],loc="best")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
a=[10,20,30,40,50,60,70,80]
b=[4,6,7,6,7,5,4,4,]
x=[1,2,3,4,8,9,1,2]
plt.scatter(a,b)
plt.scatter(a,x)
# customization
plt.scatter(a,b, c='g', s=200,edgecolors='y',marker='o',alpha =0.5)
plt.scatter(a,x,c='r', s=400,edgecolors='b',marker='4',alpha =1)
plt.legend(['b','x'],loc='best')

plt.title("Scatter PLOT VIEW")
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.grid(False)
plt.savefig("Scatter PLOT VIEW.png")
plt.show()








```

# How to create abox plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
# creating variables
table=[1,3,45,20,34,66,20,2,23,45,67]
order=[20,1,34,55,6,56,32,21,45,67,89,90,45,33]
demand=[3,4,78,90,55,66,44,33,22,12,66,13,77]
data=list([table,order,demand])
plt.boxplot(data,showmeans=True)
plt.title("   fig :Box plt ")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
label=['orange','mango','grape','papaya']
sizes=[30,87,55,56]
plt.pie(sizes,labels=label,autopct='%1.1f%%',shadow=True,startangle=90,explode=(0,0.1,0,0))

plt.title('pie chart matplot')
plt.show()

```
